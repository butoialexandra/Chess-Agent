```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def show_rewards(rewards):
    plt.plot(np.arange(rewards.size), rewards)
    plt.title('REWARDS PER EPISODE')
    plt.ylabel('REWARD')
    plt.xlabel('EPISODE NUMBER')
```

```python
def show_moves(rewards):
    plt.plot(np.arange(rewards.size), rewards)
    plt.title('NUMBER OF MOVES PER EPISODE')
    plt.ylabel('NUMBER OF MOVES')
    plt.xlabel('EPISODE NUMBER')
```

```python
def show_sarsa_vs_qlearning_rewards(rewards_sarsa, rewards_q):
    plt.figure(figsize=(7,5))
    plt.plot(np.arange(rewards_sarsa.size), rewards_sarsa)
    plt.plot(np.arange(rewards_q.size), rewards_q)
    plt.title('REWARDS PER EPISODE')
    plt.ylabel('REWARD')
    plt.xlabel('EPISODE NUMBER')
    plt.legend(['Sarsa', 'Q-learning'], loc='upper left')
```

```python
def show_sarsa_vs_qlearning_moves(rewards_sarsa, rewards_q):
    plt.figure(figsize=(7,5))
    plt.plot(np.arange(rewards_sarsa.size), rewards_sarsa)
    plt.plot(np.arange(rewards_q.size), rewards_q)
    plt.title('NUMBER OF MOVES PER EPISODE')
    plt.ylabel('NUMBER OF MOVES')
    plt.xlabel('EPISODE NUMBER')
    plt.legend(['Sarsa', 'Q-learning'], loc='upper left')
```

```python
def show_rewards_gamma(rewards_gamma1, rewards_gamma2, rewards_gamma3):
    plt.figure(figsize=(7,5))
    plt.plot(np.arange(rewards_gamma1.size), rewards_gamma1)
    plt.plot(np.arange(rewards_gamma2.size), rewards_gamma2)
    plt.plot(np.arange(rewards_gamma3.size), rewards_gamma3)
    plt.title('REWARDS PER EPISODE')
    plt.ylabel('REWARD')
    plt.xlabel('EPISODE NUMBER')
    plt.legend(['$\gamma=0.75$', '$\gamma=0.85$', '$\gamma=0.95$'], loc='upper left')
```

```python
def show_moves_gamma(rewards_gamma1, rewards_gamma2, rewards_gamma3):
    plt.figure(figsize=(7,5))
    plt.plot(np.arange(rewards_gamma1.size), rewards_gamma1)
    plt.plot(np.arange(rewards_gamma2.size), rewards_gamma2)
    plt.plot(np.arange(rewards_gamma3.size), rewards_gamma3)
    plt.title('NUMBER OF MOVES PER EPISODE')
    plt.ylabel('NUMBER OF MOVES')
    plt.xlabel('EPISODE NUMBER')
    plt.legend(['$\gamma=0.75$', '$\gamma=0.85$', '$\gamma=0.95$'], loc='upper left')
```

```python
def show_rewards_beta(rewards_beta1, rewards_beta2, rewards_beta3):
    plt.figure(figsize=(7,5))
    plt.plot(np.arange(rewards_beta1.size), rewards_beta1)
    plt.plot(np.arange(rewards_beta2.size), rewards_beta2)
    plt.plot(np.arange(rewards_beta3.size), rewards_beta3)
    plt.title('REWARDS PER EPISODE')
    plt.ylabel('REWARD')
    plt.xlabel('EPISODE NUMBER')
    plt.legend(['$\\beta=0.0005$', '$\\beta=0.00005$', '$\\beta=0.000005$'], loc='lower right')
```

```python
def show_moves_beta(rewards_beta1, rewards_beta2, rewards_beta3):
    plt.figure(figsize=(7,5))
    plt.plot(np.arange(rewards_beta1.size), rewards_beta1)
    plt.plot(np.arange(rewards_beta2.size), rewards_beta2)
    plt.plot(np.arange(rewards_beta3.size), rewards_beta3)
    plt.title('NUMBER OF MOVES PER EPISODE')
    plt.ylabel('NUMBER OF MOVES')
    plt.xlabel('EPISODE NUMBER')
    plt.legend(['$\\beta=0.0005$', '$\\beta=0.00005$', '$\\beta=0.000005$'], loc='lower right')
```

Rewards per episode using Q learning

```python
with open('q_500k.pickle', 'rb') as file:
    rewards = pickle.load(file)
show_rewards(rewards)
```

```python
with open('q_500k_beta10-4.pickle', 'rb') as file:
    beta1 = pickle.load(file)
with open('q_500k.pickle', 'rb') as file:
    beta2 = pickle.load(file)
with open('q_500k_beta10-6.pickle', 'rb') as file:
    beta3 = pickle.load(file)
show_rewards_beta(beta1, beta2, beta3)
```
