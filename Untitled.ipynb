```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def show_sarsa(rewards):
    plt.plot(np.arange(rewards.size), rewards)
    plt.title('REWARDS PER EPISODE')
    plt.ylabel('REWARD')
    plt.xlabel('EPISODE NUMBER')
```

```python
def show_sarsa_vs_qlearning(rewards_sarsa, rewards_q):
    plt.figure(figsize=(7,5))
    plt.plot(np.arange(rewards_sarsa.size), rewards_sarsa)
    plt.plot(np.arange(rewards_q.size), rewards_q)
    plt.title('REWARDS PER EPISODE')
    plt.ylabel('REWARD')
    plt.xlabel('EPISODE NUMBER')
    plt.legend(['Sarsa', 'Q-learning'], loc='upper left')
```

```python
with open('sarsa_10k_4.pickle', 'rb') as file:
    rewards = pickle.load(file)
show_sarsa(rewards)
```

```python
with open('sarsa_10k_3.pickle', 'rb') as file:
    rewards_sarsa = pickle.load(file)
with open('q_10k_3.pickle', 'rb') as file:
    rewards_q = pickle.load(file)
show_sarsa_vs_qlearning(rewards_sarsa, rewards_q)
```
